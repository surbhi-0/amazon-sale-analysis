```python
plt.figure(figsize=(7,5))
ax=sns.countplot(data=amazon,x="Country")
ax.bar_label(ax.containers[0])
plt.title("Country Wise Distribution")
plt.show()
```

```python
#from the above chart;
#we have analysed tha the maximum number of orders are from Rawanda.
```

```python
gb=amazon.groupby("Units Sold").agg({"Total Revenue":'mean',"Total Cost":'mean',"Total Profit":'mean'})
print(gb)                                    
                                     
```

```python
plt.figure(figsize=(7,6))
sns.heatmap(gb, annot=True)
plt.title("relationship between Total sold and Revenues of Products")
plt.show()
```

```python
#from the above chart we have concluded that Units Solds have a good impact on the products of Revenues.....
```

```python
gb1=amazon.groupby("Unit Cost").agg({"Total Revenue":'mean',"Total Cost":'mean',"Total Profit":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(7,6))
plt.title("relationship between Total Cost and revenues of products")
sns.heatmap(gb1, annot=True)
plt.show()
```

```python
#from the above chart we have concluded that there is less impact on Cost as aompare to Sold on the products of Revenue
```
